## IF/ELSE
Evalua una o mas condiciones, y dependiendo de su valor de verdad, realizar una accion u otra (solo dos acciones)

Sintaxis



```
if (condicion):
  si la condicion es verdadera, se ejecuta este pedazo de codigo
else:
  si la condicion es falsa, se ejecuata este pedazo de codigo
```



````python
lluvia=input("Esta lloviendo? ")
````

````python
if lluvia=="si"or lluvia=="si" or lluvia=="si, Muchisimo":
  print("Use sombrilla. Guarde la ropa ")
else:
  print("Salga tranquilo ")
````
